```python
# Form sets to hold coordinates of turns
FS_turns = set([]) # forward slash turns
BS_turns = set([]) # backward slasg turns

Intersections = set([])

class Car(object):
    def __init__(self, x, y, direction):
        self.x = x
        self.y = y
        self.direction = direction
        self.intersections = 0
        
        
    def update(self):
        """
        Move and then turn when needed. Crashes are checked before this update.
        """
        if self.direction=='>':
            self.x += 1
        elif self.direction=='v':
            self.y += 1
        elif self.direction=='<':
            self.x -= 1
        elif self.direction=='^':
            self.y -= 1
        # Check to see if we've hit a turn
        coords = (self.x,self.y)
        if coords in FS_turns:
            if self.direction=='>':
                self.direction='^'
            elif self.direction=='v':
                self.direction='<'
            elif self.direction=='<':
                self.direction='v'
            elif self.direction=='^':
                self.direction='>'
        if coords in BS_turns:
            if self.direction=='>':
                self.direction='v'
            elif self.direction=='v':
                self.direction='>'
            elif self.direction=='<':
                self.direction='^'
            elif self.direction=='^':
                self.direction='<'
        # Check to see if we've hit an intersection
        if coords in Intersections:
            if self.direction=='>':
                self.direction = ['^','>','v'][self.intersections]
            elif self.direction=='v':
                self.direction = ['>','v','<'][self.intersections]
            elif self.direction=='<':
                self.direction = ['v','<','^'][self.intersections]
            elif self.direction=='^':
                self.direction = ['<','^','>'][self.intersections]
            self.intersections = (self.intersections + 1) % 3
        

cars = []
        
# Parse all of the input (turn locations, intersections, and where cars are)
for y in range(len(lines)):
    for x in range(len(lines[0])):
        item = lines[y][x]
        if item=='/':
            FS_turns.add((x,y))
        if item=='\\':
            BS_turns.add((x,y))
        if item=='+':
            Intersections.add((x,y))
        if item=='>':
            cars.append(Car(x=x,y=y,direction='>'))
        if item=='v':
            cars.append(Car(x=x,y=y,direction='v'))
        if item=='<':
            cars.append(Car(x=x,y=y,direction='<'))
        if item=='^':
            cars.append(Car(x=x,y=y,direction='^'))
            
            
# Now just need mechanism to update all cars ... one by one ... and detect when they crash in between

def find_crashes():
    for i in range(len(cars)):
        for j in range(i+1,len(cars)):
            if cars[i].x == cars[j].x and cars[i].y == cars[j].y:
                print("crash found at (%s,%s)" % (cars[i].x, cars[i].y))
                return(True)
    return(False)
                

def run():
    while not find_crashes():
        for c in cars:
            c.update()
            
run()
```
